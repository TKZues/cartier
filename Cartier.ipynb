```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
cartier = pd.read_csv("cartier_catalog.csv")
cartier.head()
```

```python
cartier.tags.value_counts()
```

```python
cartier['str_split'] = cartier['tags'].str.split(',')
```

```python
cartier.head(10)
```

```python

    cartier['str_split'] = cartier["tags"].str.split(',')
    cartier['metal'] = cartier.str_split.str.get(0).str.strip()
    cartier['first_gem'] = cartier.str_split.str.get(1).str.strip()
    cartier['second_gem'] = cartier.str_split.str.get(2).str.strip()
    cartier['third_gem'] = cartier.str_split.str.get(3).str.strip()
    cartier['foruth_gem'] = cartier.str_split.str.get(4).str.strip()
    cartier.fillna(0, inplace = True)
    del cartier['str_split']
```

```python
cartier.head(10)
```

```python
cartier.drop(['ref', 'image', 'tags', 'title', 'description'], axis = 1, inplace = True)
```

```python
gems = pd.concat([cartier['first_gem'], cartier['second_gem'], cartier['third_gem'], cartier['foruth_gem']], axis = 0)
```

```python
gems.head()
```

```python
gems_value = gems.value_counts()[1:].to_frame()
```

```python
gems_value.reset_index(inplace = True)
```

```python
gems_value.columns = ['gem_type', 'count']
```

```python
gems_value
```

```python
plt.figure(figsize = (15,5))
sns.barplot(x = 'gem_type', y = "count", data = gems_value)
plt.xlabel("Gem Type")
plt.ylabel("Count")
plt.title("Count of Gems")
plt.xticks(rotation = 90)
plt.show()
```

```python
cartier_category_metal = cartier.groupby('categorie')['metal'].value_counts().to_frame()
```

```python
cartier_category_metal.columns = ['count']
```

```python
cartier_category_metal.reset_index()
```

```python
plt.figure(figsize = (15,7))
sns.barplot(x = 'categorie', y = 'count', hue = 'metal', data = cartier_category_metal)
plt.xlabel("Jewels")
plt.ylabel("Count")
plt.legend(loc = "upper left")
plt.show()
```
